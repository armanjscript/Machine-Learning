```
import numpy as np
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
import matplotlib.pyplot as plt
samples=1000
mu=[0.6,0.5]
covm=[[1.6,0.0],[0.0,0.6]]

X=np.random.multivariate_normal(mean=mu,cov=covm,size=samples)
ss=StandardScaler()

X_ss=ss.fit_transform(X)

rs=RobustScaler(quantile_range=(10,90))
X_rs=rs.fit_transform(X)

mms=MinMaxScaler(feature_range=(-1,1))
X_mms=mms.fit_transform(X)



plt.scatter(X[:, 0], X[:, 1], alpha=0.4,lw=0, marker='o',  c='#6A562A')
plt.scatter(X_ss[:, 0], X_ss[:, 1], alpha=0.4,lw=0,marker='o', c='#248EF3')
plt.scatter(X_rs[:, 0], X_rs[:, 1], alpha=0.4,lw=0,marker='o', c='#922BA3')
plt.scatter(X_mms[:, 0], X_mms[:, 1], alpha=0.4,lw=0,marker='o', c='#60A124')
plt.show()
```
