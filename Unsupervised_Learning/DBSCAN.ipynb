```
import numpy as np

class Dbscan:
  
  def __init__(self,Data,eps,Minpts):
    self.Data=Data
    self.eps=eps
    self.Minpts=Minpts
  
  def fit_pred(self):
    
    labels=[0]*len(self.Data)
    C=0

    for p in range(0,len(self.Data)):
      
      if not (labels[p]==0):
        continue

      Neighbors=self.regions(p)

      if len(Neighbors) < self.Minpts:
        labels[p]=-1
      
      else:
        C+=1
        self.expandCluster(labels,p,Neighbors,C)

    return labels


  def expandCluster(self,labels, p, Neighbors, C):

    labels[p]=C
    i=0

    while i < len(Neighbors):
      
      Pn=Neighbors[i]

      if labels[Pn]==-1:
        labels[Pn]=C
      
      elif labels[Pn]==0:
        
        labels[Pn]=C

        PnNeighbors=self.regions(Pn)

        if len(PnNeighbors) >= self.Minpts:
          Neighbors=Neighbors+PnNeighbors
        
      i+=1
  
  def regions(self,p):
    
    neighbors=[]

    for Pn in range(0,len(self.Data)):
      if np.linalg.norm(self.Data[p] - self.Data[Pn]) < self.eps:
        neighbors.append(Pn)
    
    return neighbors
```

**Testing**

```
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler

centers = [[1,0],[-1,1],[0,1]]
X, labels_true = make_blobs(n_samples=2000, centers=centers, cluster_std=0.6,
                            random_state=0)

X = StandardScaler().fit_transform(X)

dbscan=Dbscan(X,0.1,10)
res=dbscan.fit_pred()

print(res)
```
